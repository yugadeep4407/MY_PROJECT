```python
# One-hot encoding for selected categorical columns
df_one_hot = pd.get_dummies(df, columns=['Outlook', 'Temperature', 'Humidity', 'Wind'], drop_first=True)

# Label encoding for the 'PlayTennis' column
label_mapping = {'No': 0, 'Yes': 1}
df_label_encoded = df.replace({'PlayTennis': label_mapping})

# Display the encoded DF
print("\nDF after One-Hot Encoding:")
print(df_one_hot)

print("\nDF after Label Encoding:")
print(df_label_encoded)
```
